# 05 Графики результатов обучения
Готовим графики для статьи по сохранённым артефактам.

```python
import pandas as pd
import matplotlib.pyplot as plt

pred = pd.read_csv("../outputs/tabular/validation_predictions.csv")

plt.figure(figsize=(7,7))
plt.scatter(pred["y_true"], pred["y_pred"], alpha=0.5, s=18)
mn = min(pred["y_true"].min(), pred["y_pred"].min())
mx = max(pred["y_true"].max(), pred["y_pred"].max())
plt.plot([mn, mx], [mn, mx], "--")
plt.xlabel("Факт")
plt.ylabel("Прогноз")
plt.title("Parity plot для валидации")
plt.grid(alpha=0.3)
plt.show()
```

# 04 Графовая схема проекта
Построение графа работ/рисков и визуализация структуры.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from src.data.graph_builder import build_wbs_graph

# Загружаем таблицы узлов и связей
nodes_df = pd.read_csv("../data/raw/graph_nodes.csv")
edges_df = pd.read_csv("../data/raw/graph_edges.csv")

# Строим граф
G = build_wbs_graph(nodes_df, edges_df)

# Рисуем граф
plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=600, font_size=8)
plt.title("Графовая схема строительного проекта")
plt.show()
```

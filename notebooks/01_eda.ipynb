# 01 EDA
Ноутбук для первичного анализа реального строительного датасета.

```python
# Импортируем библиотеки для EDA
import pandas as pd
import matplotlib.pyplot as plt

# Читаем реальный датасет
df = pd.read_csv("../data/raw/real_construction_data.csv")

# Смотрим первые строки
df.head()
```

```python
# Базовая структура и пропуски
print(df.shape)
print(df.dtypes)
print((df.isna().mean()*100).sort_values(ascending=False).head(20))
```

```python
# Пример распределения целевой переменной
df["SalePrice"].hist(bins=40)
plt.title("Распределение SalePrice")
plt.show()
```
